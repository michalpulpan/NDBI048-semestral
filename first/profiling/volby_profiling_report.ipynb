```python
import pandas as pd
from pandas_profiling import ProfileReport

volby_df = pd.read_csv('../../geodata/volby/volby_sociodemo.csv', sep=';')
volby_profile = ProfileReport(volby_df, title='Výsledky voleb', correlations={
        "auto": {"calculate": False},
        "pearson": {"calculate": False},
        "spearman": {"calculate": False},
        "kendall": {"calculate": False},
        "phi_k": {"calculate": False},
        "cramers": {"calculate": False},
    })
volby_profile.to_file(output_file="volby.html")
```
