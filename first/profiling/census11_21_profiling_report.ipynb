```python
import pandas as pd
from pandas_profiling import ProfileReport

census_df = pd.read_csv('../../geodata/census11_21/data_obce_vyhl_nevyhl.csv', sep=',')
census_profile = ProfileReport(census_df, title='Census 2011_2021', correlations={
        "auto": {"calculate": False},
        "pearson": {"calculate": False},
        "spearman": {"calculate": False},
        "kendall": {"calculate": False},
        "phi_k": {"calculate": False},
        "cramers": {"calculate": False},
    })
census_profile.to_file(output_file="census11_21_profiling_report.html")
```
